**Q1. Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports,technology, food, books, etc.).**


1.   Convert text to lowercase and remove punctuaƟon.
2.   Tokenize the text into words and sentences.
3.   Remove stopwords (using NLTK's stopwords list).
4.   Display word frequency distribution (excluding stopwords).


```python
text = """Technology is evolving at an unprecedented pace. From smartphones and AI assistants like Siri and Alexa to self-driving cars and 5G networks, the advancements are mind-blowing. I love exploring new gadgets, especially those that improve productivity. In the world of tech, there's always something innovative to look forward to. It's truly an exciting time to be alive."""


# STEP 1 LOWERCASE AND REMOVE PUNCTUATION
import string
text_lower = text.lower()
text_no_punct = text_lower.translate(str.maketrans('','',string.punctuation))

# STEP 2 TOKENIZATION
from nltk.tokenize import word_tokenize,sent_tokenize
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')

wordss = word_tokenize(text_no_punct)
sentences = sent_tokenize(text)


# STEP 3 REMOVE STOPWORDS
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_words = []
for word in wordss:
  if word not in stop_words: #if we didnt find the word in teh stop_word list then it means that it is not the stop words
    filtered_words.append(word)


# STEP 4 WORD FREQUENCY DISTRIBUTION
from nltk import FreqDist
fdist = FreqDist(filtered_words)
fdist.plot(10,title="Word Frequency Distribution")
```

**Q2: Stemming and Lemmatization**


1.   Take the tokenized words from Question 1 (after stopword removal).
2.   Apply stemming using NLTK's PorterStemmer and LancasterStemmer.
1.   Apply lemmatization using NLTK's WordNetLemmatizer.
2.   Compare and display results of both techniques.




```python
from nltk.stem import PorterStemmer,LancasterStemmer,WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')

porter = PorterStemmer()
lancaster = LancasterStemmer()
lemmatizer = WordNetLemmatizer()
porter_stems = []
for word in filtered_words:
  porter_stems.append(porter.stem(word))

lancaster_stems = []
for word in filtered_words:
  lancaster_stems.append(lancaster.stem(word))

lemmatized = []
for word in filtered_words:
  lemmatized.append(lemmatizer.lemmatize(word))

print("Porter Stems:",porter_stems)
print("Lancaster Stems:",lancaster_stems)
print("Lemmatized Words:",lemmatized)
```

**Q3. Regular Expressions and Text Spliting**


1.   Take their original text from Question 1.
1.   Use regular expressions to:
  1.   Extract all words with more than 5 letters.
  2.   Extract all numbers (if any exist in their text).
  1.   Extract all capitalized words.

2.   Use text spliƫng techniques to:

  1.   Split the text into words containing only alphabets (removing digits and specialcharacters).
  2.   Extract words starting with a vowel.




```python
import re

# a. words with more than 5 letters
words_gt5 = re.findall(r'\b[a-z,A-Z]{6,}\b',text)
# b. extract numbers
numbers = re.findall(r'\b\d+\.?\d*\b',text)
# c. capitalized words
capitalized = re.findall(r'\b[A-Z][a-z]*\b',text)

#text splitting techniques
# a. words only with alphabets
alpha_only = re.findall(r'\b[A-Za-z]+\b',text)
# b. words starting with vowels
vowel_start = re.findall(r'\b[aeiouAEIOU][A-Za-z]*\b',text)


print("Words >5 letters:", words_gt5)
print("Numbers:", numbers)
print("Capitalized Words:", capitalized)
print("Alphabet-only words:", alpha_only)
print("Words starting with vowels:", vowel_start)
```

**Q4. Custom Tokenization & Regex-based Text Cleaning**


1.   Take original text from Question 1.
1.   Write a custom tokenization function that:
  1.   Removes punctuation and special symbols, but keeps contractions (e.g.,
"isn't" should not be split into "is" and "n't").
  2.   Handles hyphenated words as a single token (e.g., "state-of-the-art" remains a single token).
  1.   Tokenizes numbers separately but keeps decimal numbers intact (e.g., "3.14"
should remain as is).

2.   Use Regex Substitutions (re.sub) to:

  1.   Replace email addresses with '<EMAIL>' placeholder.
  2.   Replace URLs with '<URL>' placeholder.
  2.   Replace phone numbers (formats: 123-456-7890 or +91 9876543210) with
'<PHONE>' placeholder.



```python
# custom tokenization
def custom_tokenizer(text):
  #keep contractions and hyphenated words,tokenize numbers,keep decimals
  tokens = re.findall(r"\b\w+(?:[-']\w+)*\b|\d+\.\d+|\d+",text)
  return tokens

tokens = custom_tokenizer(text)
print("custom tokens : ",tokens)


# regex subustitutions
sample_text = """"Contact me at example@mail.com or visit https://example.com. Call +91 9876543210 or 123-456-7890."""

email_replaced = re.sub(r'\b[\w\.-]+@[\w\.-]+\.\w+\b','<EMAIL>',sample_text)
url_replaced = re.sub(r'https?://\S+|www\.\S+','<URL>',email_replaced)
phone_replaced = re.sub(r'(\+?\d{1,3}[\s-]?)?\d{3}[\s-]?\d{3}[\s-]?\d{4}','<PHONE>',url_replaced)   # (\+?\d{1,3}[\s-]?)     ?\d{3}[\s-]    ?\d{3}[\s-]     ?\d{4}

print("cleaned email text : ",email_replaced)
print("cleaned url text: ",url_replaced)
print("cleaned phone text : ",phone_replaced)
print("all cleaned text : ",phone_replaced)
```
