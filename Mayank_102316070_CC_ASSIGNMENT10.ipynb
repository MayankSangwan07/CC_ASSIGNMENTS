## Q1.
Write a unique paragraph (5-6 sentences) about your favorite topic (e.g., sports, technology, food, books, etc.).

- Convert text to lowercase and remove punctuation using `re`.
- Tokenize the text into words and sentences.
- Split using `split()` and `word_tokenize()` and compare how Python’s `split()` and NLTK’s `word_tokenize()` differ.
- Remove stopwords (using NLTK's stopwords list).
- Display word frequency distribution (excluding stopwords).

```python
text = """Technology is evolving faster than ever before. Artificial Intelligence, Robotics, and Quantum Computing are reshaping industries. Smartphones and smartwatches have become essential parts of daily life. Cloud computing offers scalable solutions to businesses. The future will be dominated by advancements like AI, 5G, and autonomous vehicles."""
print("Text : ",text)

#1 convert to lower case and remove punctuation
import re

text_lower = text.lower()
text_clean = re.sub(r'[^\w\s]','',text_lower)

print("Clean Text : ",text_clean)

#2 tokenize into words and sentences
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize,sent_tokenize
  #tokenizing
words = word_tokenize(text_clean)
sentences = sent_tokenize(text)

print("Words : ",words)
print("Sentences :  ",sentences)

#3 compare split() vs word_tokenize()
  #using split()
words_split = text_clean.split()
print("Using split() : ",words_split)
print("using word_tokenize() : ",words)
print("\nDifference: split() simply splits at spaces, word_tokenize() is smarter at handling punctuation, special characters, etc.")

#4 remove stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

  #removing stopwords
filtered_words = []
for w in words:
  if w not in stop_words:
    filtered_words.append(w)

print("Filtered Words : ",filtered_words)

#5 word frequency distribution
from nltk.probability import FreqDist
fdist = FreqDist(filtered_words)
fdist.plot(10) #top 10 word will be there in the plot
```

## Q2.
Using the same paragraph from Q1:

- Extract all words with only alphabets using `re.findall()`.
- Remove stopwords using NLTK’s stopword list.
- Perform stemming with `PorterStemmer`.
- Perform lemmatization with `WordNetLemmatizer`.
- Compare the stemmed and lemmatized outputs and explain when you’d prefer one over the other.

```python
#Extract all words with only alphabets using re.findall().
alphabets = re.findall(r'\b[a-zA-Z]+\b',text_clean)
print("ALPHABETS : ",alphabets)

filtered_alphabets = []
for alpha in alphabets:
  if alpha not in stop_words:
    filtered_alphabets.append(alpha)

print("FILTERED ALPHABETS : ",filtered_alphabets)

#perform stemming with porterstemmer
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = []
for w in filtered_alphabets:
  stemmed_words.append(stemmer.stem(w))

print("STEMMED WORDS : ",stemmed_words)

#perform lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_words = []
for w in filtered_alphabets:
  lemmatized_words.append(lemmatizer.lemmatize(w))

print("LEMMATIZED WORDS : ",lemmatized_words)
```

## Q3.
Choose 3 short texts of your own (e.g., different news headlines, product reviews).

- Use `CountVectorizer` to generate the Bag of Words representation.
- Use `TfidfVectorizer` to compute TF-IDF scores.
- Print and interpret the top 3 keywords from each text using TF-IDF.


```python
texts = ["AI is transforming the future.","Blockchain ensures secure transactions.","Quantum computing will change the world."]

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
bow = cv.fit_transform(texts)


print("FEATURES NAME : \n",cv.get_feature_names_out())
print("BAG OF WORDS : \n",bow)
print("BAG OF WORDS (TO ARRAY) : \n",bow.toarray())

print("\n")


from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(texts)

print("FEATURES NAME USING TF-IDF : \n",tfidf.get_feature_names_out())
print("TF-IDF MATRIX : \n",tfidf_matrix)
print("TF-IDF MATRIX (TO ARRAY) : \n",tfidf_matrix.toarray())

print("\n")

import numpy as np
feature_names = tfidf.get_feature_names_out()


for i,text in enumerate(texts):
  print(f"text {i+1} : {text}")
  tfidf_score = tfidf_matrix[i].toarray().flatten()
  top_indices = np.argsort(tfidf_score)[-3:][::-1]
  for idx in top_indices:
    print(f" - {feature_names[idx]} : {tfidf_score[idx]}")

```

## Q4.
Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs Blockchain).

- Preprocess and tokenize both texts.
- Calculate:
  - Jaccard Similarity using sets
  - Cosine Similarity using `TfidfVectorizer` + `cosine_similarity()`
  - Analyze which similarity metric gives better insights in your case.

```python
text1 = "Artificial Intelligence is changing how we live and work every day."
text2 = "Blockchain technology secures online transactions and changes digital trust."


def preprocess(text):
  text = text.lower()
  text = re.sub(r'[^\w\s]','',text)
  return set(text.split())

tokens1 = preprocess(text1)
tokens2 = preprocess(text2)

print(tokens1)
print(tokens2)

# Calculate:
# a) Jaccard Similarity using sets
intersection = tokens1.intersection(tokens2)
union = tokens1.union(tokens2)

jaccard_similarity = len(intersection)/len(union)
print(f"jaccard similarity : {jaccard_similarity:.4f}")

# b) Cosine Similarity:
from sklearn.metrics.pairwise import cosine_similarity

tfidf = TfidfVectorizer()
vectors = tfidf.fit_transform([text1 , text2])

cos_sim = cosine_similarity(vectors[0:1],vectors[1:2])
print(f"consine simarlirity : {cos_sim[0][0]:.4f}")


# - Cosine Similarity gives better insight because it considers term frequency and importance.
# - Jaccard Similarity only checks set overlap and ignores word importance.
```

## Q6.
Choose your own paragraph (~100 words) as training data.

- Tokenize text using `Tokenizer()` from `keras.preprocessing.text`.
- Create input sequences and build a simple LSTM or Dense model.
- Train the model and generate 2–3 new lines of text starting from any seed word you provide.

```python
train_text = """Artificial intelligence and machine learning are important fields in computer science. They help solve complex problems efficiently and automate tasks. As technology advances, these fields will continue to evolve and create new opportunities in various industries."""

#tokenize using tokenizer
from tensorflow.keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
import numpy as np

tokenizer = Tokenizer()
tokenizer.fit_on_texts([train_text])

total_words = len(tokenizer.word_index) + 1
print("TOTAL WORDS : ",total_words)

  #input sequences
input_sequences = []
for line in train_text.split('.'):
  token_list = tokenizer.texts_to_sequences([line])[0]
  print(token_list)
  for i in range(1,len(token_list)):
    n_gram_sequence = token_list[:i+1]
    input_sequences.append(n_gram_sequence)

  #padding
max_sequence_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences,maxlen=max_sequence_len,padding='pre'))

  #feature and labels
X = input_sequences[:,:-1]
y = input_sequences[:,-1]

print(X)
print('\n')
print(y)


####BUILD AND TRAIN SIMPLE MODEL
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense

model = Sequential()
model.add(Embedding(total_words,10,input_length=max_sequence_len-1))
model.add(LSTM(50))
model.add(Dense(total_words,activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy',optimizer='adam')
model.fit(X,y,epochs=200,verbose=1)



###TEXT GENERATION
seed_text = 'Technology'
next_words = 3

for _ in range(next_words):
  token_list = tokenizer.texts_to_sequences([seed_text])[0]
  token_list = pad_sequences([token_list],maxlen=max_sequence_len-1,padding='pre')
  predicted = model.predict(token_list,verbose=0)
  predicted_word = ""

  for word,index in tokenizer.word_index.items():
    if index == np.argmax(predicted):
      predicted_word = word
      break
  seed_text = seed_text + " " + predicted_word

print(seed_text)
```
